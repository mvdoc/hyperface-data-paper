```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
from glob import glob
```

```python
fns = sorted(glob('/home/vassiki/budapest_data/outputs/fmriprep/sub-sid000007/func/*.tsv'))
```

```python
dfs = [pd.read_csv(f, sep='\t') for f in fns]
```

```python
keep_columns = [
    'framewise_displacement',
    'rot_x', 'rot_y', 'rot_z',
    'trans_x', 'trans_y', 'trans_z'
]
```

```python
dfs = [df[keep_columns] for df in dfs]
```

```python
fig, ax = plt.subplots(5, 1, figsize=(12, 8), sharey=True, sharex=True)
for axx, df in zip(ax.flat, dfs):
    d = df['framewise_displacement']
    axx.plot(d)
    axx.axhline(np.nanmean(d), color='lightgray', linestyle='dashed')
```

```python
fig, ax = plt.subplots(5, 1, figsize=(12, 8), sharey=True, sharex=True)
for axx, df in zip(ax, dfs):
    d = df[[f'trans_{w}' for w in 'xyz']]
    axx.plot(d)
```

```python
fig, ax = plt.subplots(5, 1, figsize=(12, 8), sharey=True, sharex=True)
for axx, df in zip(ax, dfs):
    d = df[[f'rot_{w}' for w in 'xyz']]
    axx.plot(d)
```

```python
fig, ax = plt.subplots(5, 1, figsize=(12, 8), sharey=True, sharex=True)
for axx, df in zip(ax.flat, dfs):
    d = df['framewise_displacement']
    axx.hist(d, bins=100);
```
